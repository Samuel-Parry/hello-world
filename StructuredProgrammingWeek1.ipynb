```python
! wget -q https://raw.githubusercontent.com/PX2134/data/master/week1/speed.txt

d = np.loadtxt('speed.txt', delimiter=',',skiprows=1) 

t = d[:, 0]; v = d[:, 1] 

plt.plot(t, v)
plt.xlabel('time')
plt.ylabel('speed')
plt.title('Exercise 4')
plt.grid(True)
```

```python
dt = 0.1
N = len(t)
x = np.zeros(N)
x[0] = 0
for i in range(N-1):
    x[i+1] = x[i] + v[i]*dt
plt.plot(t, x)

plt.xlabel('time')
plt.ylabel('position')
plt.title('Exercise 4')
plt.grid(True)
```
